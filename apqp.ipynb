# [dbo].[FUNC_GET_TABLE_SECTION_PROCESSING_MASSPRODUCTION]

```sql
/****** Object:  UserDefinedFunction [dbo].[FUNC_GET_TABLE_SECTION_PROCESSING_MASSPRODUCTION]    Script Date: 8/16/2024 8:11:50 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[FUNC_GET_TABLE_SECTION_PROCESSING_MASSPRODUCTION] (@CONTROL_NO NVARCHAR(50))
RETURNS TABLE
AS
RETURN
 (
	SELECT  DISTINCT PIC_SECTION , CONTROL_NO FROM TBL_FEASIBILITY 
	WHERE CONTROL_NO = @CONTROL_NO AND ID_SITUATION != '03' 
	AND (LEN(DATE_ORDER_EQUIPMENT) = 0  or LEN(DATE_RECEIVE_EQUIPMENT) =0 or LEN(RTRIM(ID_SITUATION)) = 0 or LEN(RTRIM(DETAILS_PREPARED)) = 0)
 )
GO
```

# [dbo].[VM_GET_SECTION_AND_DUE_DATE_TABLE_FEASIBILITY]

```sql
/****** Object:  View [dbo].[VM_GET_SECTION_AND_DUE_DATE_TABLE_FEASIBILITY]    Script Date: 8/16/2024 8:11:51 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[VM_GET_SECTION_AND_DUE_DATE_TABLE_FEASIBILITY]
AS
SELECT
DISTINCT 
TBL_FEASIBILITY.PIC_SECTION,
TBL_NEW_MOLD_MST.CONTROL_NO,
TBL_NEW_MOLD_MST.FEASIBILITY,
TBL_NEW_MOLD_MST.APPROVED_FEASIBILITY ,
TBL_NEW_MOLD_MST.DUE_DATE_PRODUCT_INFORMATION,
IIF((DATE_ORDER_EQUIPMENT is null or DATE_RECEIVE_EQUIPMENT is null or LEN(RTRIM(ID_SITUATION)) = 0 or LEN(RTRIM(DETAILS_PREPARED)) = 0 ) AND ID_SITUATION NOT IN('02','03'), 'GET_SECTION_DUE_DATE' , 'PROCESSING_DUE_DATE') AS STEP_PROCESSING 
FROM TBL_FEASIBILITY
LEFT JOIN TBL_NEW_MOLD_MST
ON TBL_FEASIBILITY.CONTROL_NO = TBL_NEW_MOLD_MST.CONTROL_NO
WHERE 
  (DATE_ORDER_EQUIPMENT is null or DATE_RECEIVE_EQUIPMENT is null or LEN(RTRIM(ID_SITUATION)) = 0 or LEN(RTRIM(DETAILS_PREPARED)) = 0)
GO
```

# [dbo].[VM_GET_SECTION_AND_DUE_DATE_ONE_COLUMN_FEASIBILITY]

```sql
/****** Object:  View [dbo].[VM_GET_SECTION_AND_DUE_DATE_ONE_COLUMN_FEASIBILITY]    Script Date: 8/16/2024 8:11:51 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[VM_GET_SECTION_AND_DUE_DATE_ONE_COLUMN_FEASIBILITY]
AS
SELECT CONTROL_NO , string_agg(PIC_SECTION, ', ')  + ' \ '  + IIF(DUE_DATE_PRODUCT_INFORMATION IS NULL , '', format(DUE_DATE_PRODUCT_INFORMATION, 'dd-MMM')) AS DUE_DATE

FROM [VM_GET_SECTION_AND_DUE_DATE_TABLE_FEASIBILITY]
GROUP BY CONTROL_NO ,DUE_DATE_PRODUCT_INFORMATION

GO
```
